[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/sensein/senselab/blob/alpha/tutorials/senselab-ai/senselab_ai_intro.ipynb)

senselab-ai lets you analyze your data by chatting in natural language.
You can simply ask the agent what you want to do (e.g., load audio files, preprocess them, extract features, transcribe, etc.), and it will generate and run the corresponding senselab code while also showing you the code it used.

This makes it perfect for exploration and for users who prefer not to code directly.

## 3. Open the Settings
Inside JupyterLab, open the menu: Settings → Notebook Intelligence Settings

## 4. Configure the AI Model
In the settings dialog, select a model provider.
For this tutorial we’ll use OpenAI GPT-4o:
- Provider: OpenAI Compatible
- Model: gpt-4o
- API key: paste your OpenAI key (starts with sk-...)

⚡️ The extension also supports other providers, including Ollama for running local models (requires Ollama setup).

![Example of how senselab-ai works](<resources/Screenshot 2025-09-02 at 8.21.06 PM.png>)

## 5. Start Chatting with @senselab-ai

On the left sidebar, open Notebook Intelligence.
Type messages with the prefix @senselab-ai to interact.

For example:
```
@senselab-ai /help
```
This shows built-in instructions and example prompts.

![Example of how senselab-ai works](<resources/Screenshot 2025-09-02 at 8.24.51 PM.png>)

## 6. Example: Analyze Audio Files

Suppose you have a folder with .wav files. You can ask:

```
@senselab-ai Read the audio files in "<directory_path>", 
make them mono, resample them to 16kHz, 
extract audio features, and transcribe them using the Whisper tiny model.
```

The agent will:
1. Load the files from disk
2. Preprocess (mono + resample to 16kHz)
3. Extract features (e.g., acoustic embeddings)
4. Run transcription with Whisper Tiny
5. Show you the code it generated and the results

All without you needing to write a single line of code! 🎉
